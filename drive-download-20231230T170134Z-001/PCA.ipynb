# PCA

```python
from numpy import array
from numpy import mean
from numpy import cov
from numpy.linalg import eig
```

```python
#define matrix
A = array([[1,2],[3,4],[5,6]])
print(A)
```

```python
#Calculate the mean of each column
M = mean(A.T, axis=1)
print(M)
print(A.T)
```

```python
#center columns by substracting column means
C = A-M
print(C)
```

```python
#calculate covariance matrix of centerred matrix
V = cov(C.T)
print(V)
```

```python
#eigender composition of covariance matrix
values, vectors = eig(V)
print(vectors)
print(values)
```

```python
#project data
P = vectors.T.dot(C.T)
print(P.T)
```
